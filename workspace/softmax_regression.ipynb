```python
import torch.nn as nn
import torch.optim as optim
```

```python
torch.manual_seed(1)

x_data = [
    [1, 2, 1, 1],
    [2, 1, 3, 2],
    [3, 1, 3, 4],
    [4, 1, 5, 5],
    [1, 7, 5, 5],
    [1, 2, 5, 6],
    [1, 6, 6, 6],
    [1, 7, 7, 7]
    ]
y_data = [2, 2, 2, 1, 1, 1, 0, 0]

x_train = torch.FloatTensor(x_data)
y_train = torch.LongTensor(y_data)
# 데이터 갯수 8개, 데이터의 클래스가 3개라서 (8, 3)으로 인코딩 해야 함
y_one_hot = torch.zeros(8, 3)
y_one_hot.scatter_(1, y_train.unsqueeze(1), 1)
print(y_one_hot)
```

```python
W = torch.zeros((4, 3), requires_grad=True) # 가중치의 갯수 3, 클래스의 갯수 3
b = torch.zeros((1, 3), requires_grad=True) # 편향은 1개로 충분함, 클래스의 갯수 3
```

```python
optimizer = optim.SGD([W, b], lr=0.1)
epochs = 10000
```

```python
for epoch in range(epochs + 1):
    y_hat = F.softmax(x_train.matmul(W)+b, dim=1)
    cost = (y_one_hot * -torch.log(y_hat)).sum(dim=1).mean()

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print(
            f'Epoch: {epoch} | Cost: {cost.item()}'
        )
```

```python
y_hat = F.softmax(x_train.matmul(W)+b, dim=1)
print(y_hat)
print(y_one_hot)
```

```python
for y in y_hat:
    print(y.argmax().item())
```

```python
# 모델로 소프트맥스 회귀 구현하기
model = nn.Linear(4, 3) # 입력값 4개, 출력값 3개
optimizer = optim.SGD(model.parameters(), lr=0.1)
epochs = 10000

for epoch in range(epochs + 1):
    y_hat = model(x_train)
    # 소프트맥스의 비용 함수는 크로스 엔트로피, 로지스틱의 비용 함수는 이진 크로스 엔트로피
    cost = F.cross_entropy(y_hat, y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print(
            f'Epoch: {epoch} | Cost: {cost.item()}'
        )
```

```python
preds = model(x_train)
print(preds)
for pred in preds:
    print(pred.argmax().item())
```

```python
class SoftmaxModel(nn.Module):
    def __init__(self, in_features, out_features):
        super().__init__()
        self.linear = nn.Linear(in_features, out_features)
    
    def forward(self, x):
        return self.linear(x)
```

```python
model = SoftmaxModel(4, 3)
optimizer = optim.SGD(model.parameters(), lr=0.001)
epochs = 50000
```

```python
for epoch in range(epochs + 1):
    y_hat = model(x_train)
    # F.cross_entropy()를 사용하면 메소드 안에 소프트맥스 함수가 내장되어 있어서
    # model 안에서 nn.Softmax()를 생략하고 xx.Linear()만 불러와도 됨
    cost = F.cross_entropy(y_hat, y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    
    if epoch % 100 == 0:
        print(
            f'Epoch: {epoch} | Cost: {cost.item()}'
        )
```

```python
preds = model(x_train)
for pred in preds:
    print(pred.argmax().item())
```

```python
import sklearn.datasets as datasets
import numpy as np

iris = datasets.load_iris()
iris
```

```python
x_train = torch.FloatTensor(iris.data)
y_train = torch.LongTensor(iris.target)
print(x_train.size())
print(y_train.size())
```

```python
model = SoftmaxModel(iris.data.shape[1], len(np.unique(iris.target)))
optimizer = optim.SGD(model.parameters(), lr=0.001)
epochs = 50000

for epoch in range(epochs + 1):
    y_hat = model(x_train)
    cost = F.cross_entropy(y_hat, y_train)
    
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    
    if epoch % 100 == 0:
        print(
            f'Epoch: {epoch} | Cost: {cost.item()}'
        )
```

```python
pred = model(x_train)
print(pred)
```

```python
accuracies = []
for i in range(len(pred)):
    accuracies.append(y_train[i] == pred[i].argmax())

print(torch.FloatTensor(accuracies).sum() / len(torch.FloatTensor(accuracies)))
```
